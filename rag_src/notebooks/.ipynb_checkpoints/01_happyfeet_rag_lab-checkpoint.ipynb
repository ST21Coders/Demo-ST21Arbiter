# HappyFeet RAG Lab &nbsp;·&nbsp; 01 — End-to-End DIY Pipeline

**The reusable notebook.** It builds a complete Retrieval-Augmented Generation system for
**HappyFeet**, a fictional US shoe retailer with 80+ outlets, using only:

| Concern | Choice |
|---|---|
| Embeddings | Amazon **Titan Text Embeddings v2** |
| Vector store | **Amazon S3 Vectors** |
| Reranking | **Bedrock Rerank** (Amazon / Cohere) |
| Generation | **Claude Sonnet** (swap in one config line) |

```
[HR PDFs] -> chunk -> Titan embed -> [S3 Vectors index]
                                          |
         query -> Titan embed ------------+--> top-k -> rerank -> Claude Sonnet -> answer
```

> The same `happyfeet_rag` library powering this notebook powers the deployed **AgentCore**
> chatbot (`agent/`) — so what you validate here is exactly what ships. Nothing is duplicated.

**Prerequisites:** `pip install -e ".[data,notebook,dev]"`, AWS credentials, and Bedrock model
access for Titan v2 + Claude Sonnet in your region. Run the preflight cell first — do not skip it.

## 0 · Imports and configuration

Everything resolves from `config/settings.toml`. To change the LLM, edit `generation_model_id` there (or set `BEDROCK_GENERATION_MODEL_ID`).

````python
%load_ext autoreload
%autoreload 2
import json, pathlib, datetime as dt
import happyfeet_rag
from happyfeet_rag.config import get_settings
from happyfeet_rag import preflight, loaders, chunking, embeddings, vectors, rerank, generation, retrieval, evaluation

S = get_settings()            # active environment (HAPPYFEET_ENV, default 'dev')
REPO = pathlib.Path(happyfeet_rag.config.REPO_ROOT)
print(json.dumps(S.summary(), indent=2))
````

## 1 · Preflight — do not skip

Checks boto3/s3vectors, credentials, region, and Bedrock model access without mutating anything.

````python
READY = preflight.run_preflight(S)
````

````python
# Guard: the AWS cells below assume preflight passed. If it didn't, fix the FAIL rows and re-run.
if not READY:
    raise SystemExit('Preflight not ready — configure AWS creds / region / Bedrock model access, then re-run from here.')
````

## 2 · Generate the mock corpus (idempotent)

Deterministic HR policy PDFs (unstructured) + a sales workbook (structured). Safe to re-run.

````python
hr_dir = REPO / 'data' / 'generated' / 'hr_policies'
if not hr_dir.exists() or not list(hr_dir.glob('*.pdf')):
    import subprocess
    subprocess.run([__import__('sys').executable, str(REPO/'data'/'generators'/'gen_all.py')], check=True)
print('HR PDFs:', sorted(p.name for p in hr_dir.glob('*.pdf')))
````

## 3 · Load and chunk the HR documents (UNSTRUCTURED path)

Chunking is the single biggest lever on RAG quality. The strategy comes from config
(`chunking.strategy`); notebook **02** compares fixed vs semantic head-to-head.

````python
docs = loaders.iter_hr_documents(hr_dir)
print(f'{len(docs)} HR documents loaded')

def to_epoch_days(iso):
    return (dt.date.fromisoformat(iso) - dt.date(1970,1,1)).days if iso else 0

all_chunks = []
for d in docs:
    base_meta = {
        'policy_category': d['policy_category'], 'state': d['state'], 'access_level': d['access_level'],
        'title': d['title'], 'source_uri': d['source_uri'],
        'effective_date': d['effective_date'], 'effective_epoch': to_epoch_days(d['effective_date']),
    }
    all_chunks += chunking.chunk_document(
        d['text'], d['doc_id'], strategy=S.chunk_strategy,
        max_chars=S.chunk_max_chars, overlap_chars=S.chunk_overlap_chars,
        chunking_version=S.chunking_version, metadata=base_meta,
    )
print(f'{len(all_chunks)} chunks using strategy={S.chunk_strategy!r}')
print('sample chunk id:', all_chunks[0].id)
print(all_chunks[0].text[:300], '...')
````

## 4 · Embed with Titan v2

**The golden rule:** use the *same* embedding model + dimension for documents and queries. `get_settings()` guarantees it.

````python
probe = embeddings.embed_text('How much paid time off do I get?', S)
print('query embedding dim:', len(probe), '(expected', S.embedding_dim, ')')

chunk_vecs = embeddings.embed_texts([c.text for c in all_chunks], S)
print('embedded', len(chunk_vecs), 'chunks')
````

## 5 · Create the S3 Vectors bucket + HR index

**The index schema is IMMUTABLE.** Dimension, distance metric, and the non-filterable key set
are fixed at creation. Large text (`chunk_text`) is non-filterable; everything else stays
filterable so the agent can scope retrieval by `policy_category`, `state`, `access_level`, etc.

````python
vx = vectors.make_client(S.region)
vectors.ensure_vector_bucket(vx, S.vector_bucket_name)
vectors.ensure_index(
    vx, S.vector_bucket_name, S.hr_index_name,
    dimension=S.embedding_dim, distance_metric=S.distance_metric,
    non_filterable_keys=vectors.HR_NON_FILTERABLE_KEYS,
)
print('bucket:', S.vector_bucket_name, '| index:', S.hr_index_name)
print('non-filterable keys:', vectors.HR_NON_FILTERABLE_KEYS)
````

## 6 · Ingest — `put_vectors` with deterministic ids + metadata

Each vector's key is a deterministic hash of `(doc_id, chunking_version, text)`, so re-running
this cell is idempotent and a document's vectors can be deleted later (right-to-be-forgotten).

````python
records = [
    {'key': c.id, 'embedding': v, 'metadata': {**c.metadata, 'chunk_text': c.text}}
    for c, v in zip(all_chunks, chunk_vecs)
]
written = vectors.put_records(vx, S.vector_bucket_name, S.hr_index_name, records, batch_size=S.ingest_batch_size)
print('ingested', written, 'vectors into', S.hr_index_name)
````

## 7 · Retrieve — semantic search, then a metadata-filtered search

Note the top distance printed on each hit: alert when the best match is still far away ("no good answer in the corpus").

````python
clients = retrieval.Clients.build(S)
q = 'How many paid sick days do full-time employees receive?'
for c in retrieval.retrieve(q, S.hr_index_name, S, clients=clients, use_rerank=False):
    print(f'{c.distance:.4f}  {c.doc_id:14} {c.text[:90]}')
````

````python
# Access control / precision: filter to only 'leave' policy chunks.
flt = vectors.build_filter(equals={'policy_category': 'leave'})
for c in retrieval.retrieve('How much parental leave is paid?', S.hr_index_name, S,
                            metadata_filter=flt, clients=clients, use_rerank=False):
    print(f'{c.distance:.4f}  {c.metadata.get("policy_category"):12} {c.text[:90]}')
````

## 8 · Rerank — before vs after

Vector similarity ≠ relevance. Retrieve 20 candidates, rerank to the best 5 with a cross-encoder.

````python
q = 'What discount do employees get when buying shoes?'
print('--- vector order (no rerank) ---')
for c in retrieval.retrieve(q, S.hr_index_name, S, clients=clients, use_rerank=False):
    print(f'{c.distance:.4f}  {c.doc_id:14} {c.text[:80]}')
print('\n--- reranked order ---')
for c in retrieval.retrieve(q, S.hr_index_name, S, clients=clients, use_rerank=True):
    print(f'score={c.rerank_score:.4f}  {c.doc_id:14} {c.text[:80]}')
````

## 9 · Generate a grounded answer with Claude Sonnet

Guardrails (if configured) screen input and output. Swap the model by editing `generation_model_id` in settings.toml — no code change.

````python
ans = retrieval.answer('How is sales commission calculated and when is it paid?', S.hr_index_name, S, clients=clients)
print(ans.answer)
print('\n--- grounded on ---')
for c in ans.contexts:
    print(' -', c.doc_id, '|', c.metadata.get('title'))
print(f'\ntokens: in={ans.input_tokens} out={ans.output_tokens}  model={S.generation_model_id}')
````

## 10 · Evaluate retrieval against the golden set

Cheap, deterministic, no-LLM feedback. Run this every time you touch chunking or embeddings.
Answer-quality (LLM-as-judge) and the improvement loop live in `eval/` and `docs/evaluation.md`.

````python
cases = [json.loads(l) for l in (REPO/'eval'/'golden'/'hr_qa.jsonl').read_text().splitlines() if l.strip()]
per = []
for case in cases:
    ctxs = retrieval.retrieve(case['question'], S.hr_index_name, S, clients=clients, use_rerank=True)
    per.append({
        'retrieved_ids': [c.doc_id for c in ctxs],
        'relevant_ids': case['relevant_doc_ids'],
        'top_distance': ctxs[0].distance if ctxs else None,
    })
report = evaluation.aggregate_retrieval(per, S.retrieval_top_k)
print(report)
````
